# EDA

```python
sz=df.shape
print("There are {0} records and {1} features".format(sz[0],sz[1]))
```

```python
def dist_plot(dataframe):
    """
        Plot the distribution of all the features in the dataframe
        
        @Parameter: 
        dataframe ((pd.dataframe): Name for the dataframe with the features (pd.dataframe)
        
    """
    for col in dataframe.columns:
        dist=sns.displot(dataframe, x=col)
        plt.savefig('output/{}_distribution.png'.format(col))
        
dist_plot(df)
#Cabin is missing lots of values based on distribution
```

```python
df1=df[['SibSp','Parch','Age','Sex','Pclass','Fare','Survived']]
```

```python
df1.isnull().sum()
```

```python
column_names=['Age']

def med_impute(col_names_arr,dataframe):
    """
        Impute missing values of features using the median value
        
        @P: 
        col_names_arr (list) : list of the column names that need to be normalized
        dataframe: the dataframe with the columns ref in the array and to have normalized dat appended        
    """
    
    for c in col_names_arr:
        dataframe[c].fillna(int(dataframe[c].median()), inplace=True)

med_impute(column_names,df1)
```

```python
df1['Fam_Sz']=df1['SibSp']+df1['Parch'] #A large family may have a lower chance of surviving.
```

```python
df1['child_elder_flag']=np.where((df['Age']<18) & (df['Age']>65), 1, 0) #Children and older adults may struggle to quickly leave the ship.
```

```python
df1['not_high_clss_flag']=np.where(df['Pclass']!=3, 1, 0) #Higher class passengers may have more means to leave the boat
```
